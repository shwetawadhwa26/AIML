```python
import numpy as np
data.columns
data['workclass'] = data['workclass'].replace('?',np.nan)
data['occupation'] = data['occupation'].replace('?',np.nan)
data['native-country'] = data['native-country'].replace('?',np.nan)
data.isin(['?']).sum()
data.isnull().sum()

data.isnull()
sns.heatmap(data.isnull())
```

```python
#percent of null values in every column
percent_age_of_missing_value = data.isnull().sum()*100/len(data)
percent_age_of_missing_value
```
